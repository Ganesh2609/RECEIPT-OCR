```python
import torch
from torch import nn 
from torch.utils.data import DataLoader
from torchvision import transforms

from Restformer import Restformer
from dataset import DocumentDenoisingData
from trainer import train_models

import matplotlib.pyplot as plt
from torchinfo import summary
from pathlib import Path
```

```python
# device agnostic code 

if torch.cuda.is_available():
    device = 'cuda'
else:
    device='cpu'
device
```

```python
# Hyperparameters 

IMG_SIZE = 128
MODEL_SAVE_PATH = 'Models/first_train.pth'
RESULT_SAVE_PATH = 'Results/'
```

```python
# Setting up the transforms 

input_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Grayscale(),
    transforms.Resize(size=(IMG_SIZE,IMG_SIZE)),
    transforms.Normalize(mean=[0.5], std=[0.5])
])

target_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Grayscale(),
    transforms.Resize(size=(IMG_SIZE,IMG_SIZE)),
    transforms.Normalize(mean=[0.5], std=[0.5])
])
```

```python
# Dataset and dataloaders 

root_dir = 'data'
data = DocumentDenoisingData(root_dir=root_dir, transform=input_transform, target_transform=target_transform)
```

```python
# Creating model instance

model = Restformer().to(device)
model_file = Path(MODEL_SAVE_PATH)

if model_file.is_file():
    model.load_state_dict(torch.load(f=MODEL_SAVE_PATH))
    print("Model exists...")
else:
    print("Model created...")
```

```python
# visualising random samples and predictions

rnd_idx = torch.randint(0, len(data), (1,))
rnd_data = data[rnd_idx]

input_img = rnd_data['Input'].to(device)
target_img = rnd_data['Target'].to(device)

with torch.inference_mode():
    predicted_img = model(input_img.unsqueeze(dim=0)).squeeze(dim=0)

plt.figure(figsize=(10, 5))

plt.subplot(1,3,1)
plt.imshow(input_img.permute(1,2,0).cpu(), cmap='gray')
plt.title('Input img')
plt.axis(False);

plt.subplot(1,3,2)
plt.imshow(target_img.permute(1,2,0).cpu(), cmap='gray')
plt.title('Target img')
plt.axis(False);

plt.subplot(1,3,3)
plt.imshow(predicted_img.permute(1,2,0).cpu(), cmap='gray')
plt.title('Denoised img')
plt.axis(False);
```
