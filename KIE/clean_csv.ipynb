```python
import csv
import os

def clean_csv(file_path, cleaned_path, expected_columns=9):
    with open(file_path, 'r') as infile, open(cleaned_path, 'w', newline='') as outfile:
        reader = csv.reader(infile)
        writer = csv.writer(outfile)

        for row in reader:
            # If the row has more columns than expected, merge the extras
            if len(row) > expected_columns:
                row = row[:expected_columns - 1] + [','.join(row[expected_columns - 1:])]
            writer.writerow(row)
        print(f"Cleaned CSV written to: {cleaned_path}")
```

```python
original_csv = 'sroie-2019/raw/box'
cleaned_csv = 'sroie-2019/raw/box cleaned'

for file in os.listdir(original_csv):
    clean_csv(os.path.join(original_csv, file), os.path.join(cleaned_csv, file))
```
